```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('bestsellers with categories.csv')
df.head()
```

# NO 1

The best selling books of the decade

```python
plt.rcParams['figure.figsize'] = (20,5)
df[['Name', 'Reviews']].nlargest(15, 'Reviews').plot.bar(x = 'Name', width = 0.8)
plt.title('15 Best selling Books of the Decade according to reviews', fontsize = 20)
plt.xlabel('Name of Books', fontsize = 20)
plt.ylabel('No of Reviews', fontsize = 20)
```

# No 2

THE PERCENTAGE OF FICTION TO NON-FICTION

```python
df['Genre'].value_counts()
```

```python
Non_Fiction = df['Genre'].value_counts()[0]/len(df)
Non_Fiction * 100
```

```python
Fiction = df['Genre'].value_counts()[1]/len(df)
Fiction * 100
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10,5)

df['Genre'].value_counts().sort_values(ascending = False).plot.bar(color = 'black')
plt.title('No of Published Genre over the decade', fontsize = 25)
plt.xlabel('GENRE', fontsize = 20)
```

# NO 3

Trend Over the years

```python
plt.rcParams['figure.figsize'] = (20,5)
df[['Name', 'Year']].nlargest(15, 'Year').plot.bar(x = 'Name', width = 0.8, color = 'black')
plt.title('The Trend of sales of books according to years', fontsize = 20)
plt.xlabel('Name of Books', fontsize = 20)
plt.ylabel('Years', fontsize = 20)
```

```python
df['Year'].value_counts()
```

# NO 4

Top 10 Authors according to reviews

```python
plt.rcParams['figure.figsize'] = (20,5)

df[['Author', 'Reviews']].nlargest(10, 'Reviews').plot.bar(x = 'Author', y = 'Reviews', width = 0.8)
plt.title('Top 10 Authors according to reviews', fontsize = 20)
plt.xlabel('Name of Authors', fontsize = 20)
plt.ylabel('No of Reviews', fontsize = 20)
```
